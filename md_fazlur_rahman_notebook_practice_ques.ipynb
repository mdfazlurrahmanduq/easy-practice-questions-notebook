## Question_1 
Write a Python program which iterates the integers for a given start, stop range of
values (inclusive). For numbers that are multiples of 7 print
"Bert" instead of the number. For numbers that are multiples of 11 print "Ernie".
For numbers which are multiples of both 7 and 11 print "Bert and Ernie".

For all other numbers, print the number **only** if the number is a prime number.  If the number is
not a prime number, print an asterisks (\*).
Recall, a prime number is any number that has *exactly* 2 unique factors.

### Deliverables
Author your solution using the test data provided in the code-cell below.

````python
#TEST DATA
start_value = 1
stop_value = 997
resolved_list = []
for number in range(start_value,stop_value+1):
    if number%7 == 0 and number%11 == 0:
        number = "Bert and Ernie"
    elif number%7 == 0:
        number= "Bert"
    elif number%11 ==0:
        number ="Ernie"
    else:
        if number>1:
            for i in range(2,number):
                if int(number)%i ==0:
                    number = " * "
                    break
        elif number==1:
             number = " * "
    resolved_list.append(number)
print(resolved_list)
````

## Question_2 
In Reverse Oompa-Loompa notation (RON), the operators follow their operands.
For example, to add 3 and 4, one would write 3 4 + rather than 3 + 4.
If there are multiple operations, operators are given immediately
after their second operands; so the expression 3 - 4 + 5
in conventional notation would be written as 3 4 - 5 + in RON.
Following the RON algorithm, 4 is first subtracted from 3, then 5 is added to the result.
<br/>

### Program Requirements
Author a program that:
1. Invokes a **UDF** designed to compute and return the result for a given RON input string.
2. Displays the value returned by the function.

### Function Requirements
Author a function to correctly parse a simple RON-like input string.
Your function must:
1. Accept a string as an argument
2. Calculate the RON result using the operators and digits appearing in the
input string. Your function should correctly interpret and apply **ONLY** the following 3
mathematical operators: *, +, -. See the example below.
3. If the argument string contains digits but does not contain
operators, your function should return the number
 indicated by only the digits. Ex: 1 2 should be returned as the number 12.
4. return the mathematical result of the RON calculation

**Function Example:**
- *Given Input:* 11 1 - 5 + 3 *
- *Expected Output:* 45
- *Explanation:* ((11-1)+5)\*3. Note that an operation is performed
upon encountering a operator. Ie, 11 - 1 is carried as a result of
encountering the subtraction operator '-'. 5 is then added to the previous result upon encountering the
addition operator '+'. Lastly, 3 is multiplied by the previous result upon encountering the
multiplication operator *.

**Note**: You may assume there are no errors in the
RON specification and that a single space character separates operands and operators as
shown in the example above.  In addition, you may assume all operands are interpretable
as integer (ie, there are no floating point numbers).

Author your solution using the test data provided in the code-cell below.

````python
# MODIFY THE FOLLOWING FUNCTION AS REQUIRED BY THE PROBLEM STATEMENT
def ron_calculator(eqn :str) -> int:
    list_ = []
    values = eqn.split(' ')
    for value in values:
        if value.isnumeric():
            list_.append(value)
        elif value == "*":
            num_1st = int(list_.pop())
            num_2nd = int(list_.pop())
            result = num_2nd*num_1st
            list_.append(str(result))
        elif value == "+":
            num_1st = int(list_.pop())
            num_2nd = int(list_.pop())
            result = num_2nd+num_1st
            list_.append(str(result))
        elif value == "-":
            num_1st = int(list_.pop())
            num_2nd = int(list_.pop())
            result = num_2nd-num_1st
            list_.append(str(result))
        else:
            continue
    
    if len(list_)==1:
        return int(list_[0])
    else:
        a_ans = ""
        for number in list_:
            a_ans+= number
        return int(a_ans)
                
            

# TEST DATA

# CALCULATE THE RPN RESULT FOR EACH OF THE FOLLOWING STRINGS
str_1 = '30 1 - 9 + 2 *' # Answer s.b. 76 ((30-1)+9)*2
str_2 = '4 2 * 3 * 6 -'    # Answer s.b. 18 ((4*2)*3)-6
str_3 = '40 2 3 6'       # Answer s.b. 40236 (there are no ops)

# USE/INVOKE YOUR MODIFIED ron_calculator FUNCTION AS REQUIRED BY THE PROBLEM STATEMENT
ans_1 = ron_calculator(str_1)
ans_2 = ron_calculator(str_2)
ans_3 = ron_calculator(str_3)
print(f'Result of Expression({str_1}) = {ans_1}')
print(f'Result of Expression({str_2}) = {ans_2}')
print(f'Result of Expression({str_3}) = {ans_3}')
````

### Problem Statement
Using the Grant's finch beak data, estimate and display an interval estimate for the
population proportion of finch beaks that
will fall **within** 1.5 standard deviations of the mean 95% of the time.

### Deliverables
Author your solution using the test data provided in the code-cell below.

````python
import numpy as np
from math import sqrt
import scipy.stats as stats

# TEST DATA
in_file = 'beak_depth_2012_v1.txt'
b_depth = []
with open(in_file) as fh:
    for estimation in fh:
        estimation = estimation.rstrip()
        if len(estimation)==0:
            continue
        b_depth.append(float(estimation))
b_depth_arr = np.array(b_depth)
n = len(b_depth_arr)
df = n-1
alpha = .05
mean = b_depth_arr.mean()
sd = b_depth_arr.std(ddof=1)
z_stat= stats.norm.ppf(1 - alpha/2)
p_filter = np.logical_and( b_depth_arr >= (mean - 1.5*sd), b_depth_arr <= (mean + 1.5*sd))
b_depth_arr_filtered= b_depth_arr[p_filter]
p=len(b_depth_arr_filtered)/n

p_upper= p+ (z_stat*sqrt(p*(1-p)/n))
p_lower= p- (z_stat*sqrt(p*(1-p)/n))

print(f'Interval estimate for population proportion within 1.5 SD of mean:')
print(f'Lower is {p_lower:.4f},\nand Upper is {p_upper:.4f} ')
````

## Question_4 
Every year for the past 40-plus years, Peter and Rosemary Grant have
gone to the Galápagos island of Daphne Major and collected
data on Darwin's finches. Measurements taken include beak depth and length.

### Problem Statement
The data file provided contains beak length measurements for two species of finch -
fortis and scandens. The species do not interbreed and thus, the populations are considered
independent.  Conduct a two sample test of beak means under the null hypothesis that the
inter-species beak lengths are the same.

Since you understand the risks of False Positives associated with hypothesis testing, you want to
be sure that your test rejects a true H0 no more than 5% of the time.

### Data File Format
There is a single data file that contains beak length measurements for both species of finch.

Each line in the file takes the following form:
```
species,beak_length
```
*Example:*:
```
fortis,10.6
fortis,10.6
scandens,14.3
scandens,12.5
```

Once you have removed an input line's trailing newline character,
you can split the line text into 2 variables as follows.
```
species_str, beak_len_str = line.split(',')
# The beak length values may be safely converted to floating point numbers
```
### Deliverables
1. Clearly state your null and alternative hypotheses
as comments in the code-cell as (and where) indicated by my instructional comments.
2. Clearly state your assessment/interpretation of your results (ie, reject H0, etc.)
as comments in the code-cell as (and where) indicated by my instructional comments.

````python
# YOUR NULL AND ALTERNATIVE HYPOTHESES GO HERE
# H0: Status quo: Mean is same across inter-species
# HA: Mean is not same across inter-species

# YOUR INTERPRETATION OF YOUR RESULTS GO HERE
# Assess the two t-stat values. If our sample t >= critical t (falls to the right into the
# region of rejection) the reject H0. Otherwise, fail to reject H0
import numpy as np
from math import sqrt
import scipy.stats as stats

# TEST DATA
in_file = 'beak_length_2012_v1.txt'

fort_list = []
scan_list = []
def open_file(f_name):
    with open(f_name) as fh:
        for line in fh:
            line = line.rstrip()
            if line.startswith('fortis'):
                species_str, beak_len_str = line.split(',')
                fort_len= float(beak_len_str)
                fort_list.append(fort_len)
            if not line.startswith('fortis'):
                species_str, beak_len_str = line.split(',')
                scan_len= float(beak_len_str)
                scan_list.append(scan_len)
                continue
open_file(in_file)
fort_arr = np.array(fort_list)
scan_arr = np.array(scan_list)

# establish alpha
alpha = .05

fort_n= len(fort_arr)
fort_df = fort_n-1
fort_mean= fort_arr.mean()
fort_var= fort_arr.var(ddof=1)

scan_n= len(scan_arr)
scan_df = scan_n-1
scan_mean= scan_arr.mean()
scan_var= scan_arr.var(ddof=1)

pool_df= (fort_n+scan_n)-2

# first, calculate the pooled variance using the formula above
vpool = (fort_df*fort_var + scan_df*scan_var) / (pool_df)

t_sample = abs((fort_mean - scan_mean)/sqrt(vpool/fort_n + vpool/scan_n ))

# Return the t-stat for df where the prob of exceeding t is .025
# Said differently, the area under the curve to the right of the t will be .025
t_critical = stats.t.ppf(1 - alpha/2, pool_df)

if t_sample >= t_critical:
    print(f'The Null Hypothesis is rejected in favor of the Alternative Hypothesis.')
    print(f'The sample means are significantly different (at the 5% level).')
    print(f'Sample t-stat({t_sample}). Critical value t-stat({t_critical}).')
else:
    print(f'The Null Hypothesis cannot be rejected.')
    print(f'The sample means are NOT significantly different (at the 5% level).')
    print(f'Sample t-stat({t_sample}). Critical value t-stat({t_critical}).')


````

````python
# YOUR NULL AND ALTERNATIVE HYPOTHESES GO HERE
# H0: Status quo: Nothing changed, Mean of Period 1 and period 2 are the same
# HA: Something changed, Mean of period 1 and period 2 are not same

# YOUR INTERPRETATION OF YOUR RESULTS GO HERE
# Assess the two t-stat values. If our sample t >= critical t (falls to the right into the
# region of rejection) the reject H0. Otherwise, fail to reject H0
import numpy as np
from math import sqrt
import scipy.stats as stats

# TEST DATA
in_file = 'us_county_paired_covid_v1.txt'

period_1_list = []
period_2_list = []

def open_file(f_name):
    with open(f_name) as fh:
        for line in fh:
            period_1_str, period_2_str = line.split(',')
            period_1_len= float(period_1_str)
            period_1_list.append(period_1_len)
            period_2_len= float(period_2_str)
            period_2_list.append(period_2_len)
            continue

open_file(in_file)

period_1_arr= np.array(period_1_list)
period_2_arr= np.array(period_2_list)

# establish alpha
alpha = .05

n= len(period_1_arr)
df= n-1
diff_lst = []
for p_1, p_2 in zip(period_1_arr,period_2_arr):
    diff_lst.append(p_1-p_2)

# calc the stats we need
diff_a = np.array(diff_lst, dtype='f')
dmean = diff_a.mean()
dstd = diff_a.std(ddof=1)

# since H0 is  Status quo: Nothing changed then the Mu(period_1) - Mu(period2) drops out as it will be zero
t_sample = abs(dmean/(dstd/sqrt(n)))

# Return the t-stat for df where the prob of exceeding t is .025
# Said differently, the area under the curve to the right of the t will be .025
t_critical = stats.t.ppf(1 - alpha/2, df)

mean_1 = sum(period_1_arr)/len(period_1_arr)
mean_2 = sum(period_2_arr)/len(period_2_arr)

print(f'Mean of Period 1: {mean_1}. Mean of Period 2: {mean_2}. Mean of Differences: {dmean}.')
if t_sample >= t_critical:
    print(f'The Null Hypothesis is rejected in favor of the Alternative Hypothesis.')
    print(f'The difference in the sample values are significantly different (at the 5% level).')
    print(f'Sample t-stat({t_sample}). Critical value t-stat({t_critical}).')
else:
    print(f'The Null Hypothesis cannot be rejected.')
    print(f'The difference in the sample values are NOT significantly different (at the 5% level).')
    print(f'Sample t-stat({t_sample}). Critical value t-stat({t_critical}).')

````
